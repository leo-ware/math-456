**In each case, sketch the set of points determined by the given condition:**

**(a) $|z - 1 + i| = 1$**

$|z - (1 - i)| = 1$

It's a circle around $(1 - i)$ with radius $1$.

In this figure, it's points on the red line.



(b) $|z + i| \le 3$

$|z - (0 - i)| \le 3$

It's the closed set inside the circle centered at $-i$ with radius $3$.

In this picture, it's points on the line and their convex hull.



(c) $|z - 4i| \ge 4$

$|z - (0 + 4i)| \ge 4$

It's the closed set which is the entire complex plane except for the points inside the circle centered at $4i$ with radius 4.

In this picture, it's the points on the line and all the points outside their convex hull.
