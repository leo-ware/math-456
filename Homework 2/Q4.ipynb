**In each case, sketch the closure of the set.**

**(a) $-\pi \lt \operatorname{arg}z \lt \pi$ for $z \ne 0$**

This is the open set of every point except the points along the negative x-axis, including zero. So, the boundary points here are, for any real $\psi \ge 0$ and integer $n$,

$$
\psi \exp(i\pi(1 + 2n))
$$

So, the closure is $\mathbb{C}$.

**(b) $|\operatorname{Re}z| \lt |z|$**

$$
|\operatorname{Re}z| \lt |z|
$$

Since both sides are positive reals,

$$
(\operatorname{Re}z)^2 \lt |z|^2
$$

But, $|z|^2 = (\operatorname{Re}z)^2 + (\operatorname{Imag}z)^2$. So,

$$
(\operatorname{Re}z)^2 \lt (\operatorname{Re}z)^2 + (\operatorname{Imag}z)^2
$$

So,

$$
0 \lt (\operatorname{Imag}z)^2
$$

$$
|\operatorname{Imag}z| \gt 0
$$

This is the open set of every complex number except the x-axis. The x-axis is the set of boundary points. So, the closure with be $\mathbb{C}$.

**(c) $\operatorname{Re}(1/z) \le 1/2$**

$$
\operatorname{Re}(1/z) \le {1 \over 2}
$$

If $z = 0$, then $1/z$ is not defined. So, $z = 0$ would be excluded from the set. Under the assumption that $z \ne 0$, we can proceed with,

$$
\operatorname{Re}(\bar{z}/|z|^2) \le {1 \over 2}
$$

$$
{\operatorname{Re}z \over |z|^2} \le {1 \over 2}
$$

Trivially, all complex numbers with negative real parts are in the set. We can further simplify.

$$
2\operatorname{Re}z \le |z|^2
$$

If $x$ and $y$ are real numbers such that $x + yi = z$, then,

$$
2x \le x^2 + y^2
$$

$$
x^2 - 2x + y^2 \ge 0
$$

$$
x^2 - 2x + 1 - 1 + y^2 \ge 0
$$

$$
(x - 1)^2 + y^2 \ge 1
$$

This is the entire complex plane except for the open circle of radius one centered at $(1, 0)$. 

The only boundary point of this set is origin. So, we can create the closure of this set by including it.

**(d) $\operatorname{Re}(z^2) \gt 0$**

Let $x$ and $y$ be real numbers such that $x + yi = z$. Then,

$$
\operatorname{Re}(z^2) = \operatorname{Re}((x^2 - y^2) + 2xyi)
$$

$$
\operatorname{Re}(z^2) = x^2 - y^2
$$

So, the set is decribed by,

$$
x^2 - y^2 \gt 0
$$

$$
x^2 \gt y^2
$$

$$
|x| \gt |y|
$$

Or,

$$
|\operatorname{Re}(z)| > |\operatorname{Imag}(z)|
$$

Consider the lines $x = y$ and $x = -y$. This is the set of points either above both or below both. The set does not include zero.

So, the boundary points are the origin and the two boundary lines.
