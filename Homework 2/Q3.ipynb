**1. Sketch the following sets and determine which are domains:**

**(a) $|z− 2 + i| ≤ 1$**

Equivalent to,

$|z− (2 - i)| ≤ 1$

It's the closed circle or radius 1 around $(2 - i)$.

This is not a domain because it's closed.

**(b) $|2z + 3| > 4$**

Let,

$$
w = 2z + 3
$$

So,

$$
|w| > 4
$$

This is the open set of all points outside the circle or radius 4 centered at the origin.

$$
z = {w \over 2} - {3 \over 2}
$$

So, imagine taking that whole set, shrinking the size of the hole in the center by a factor of 2, then translating it 3/2 units to the left. So, $z$ is the open set of points outside the circle of radius 2 centered at -3/2.

It's the points outside the circle.



This is a domain because it's open, connected, and nonempty.

**(c) $\text{Im}(z) \gt 1$**

This is the closed set of points above but not including the line $y = 1$.

This is a domain.

**(d) $\text{Im}(z) = 1$**

It's all the points along the line $y = 1$. It's the points on the red line extending to infinity in either direction.



This is not a domain because it's not open. It's not open because every point in the set is boundary point.

**(e) $0 \le \text{arg}(z) \le \pi/4$ $(z \ne 0)$**

It's all the points above (and including) the x-axis but below (or including) the line $y = x$. The origin is excluded.

This is not a domain because it's closed.

**(f) $|z - 4| \ge |z|$**

Let $x$ and $y$ be real numbers such that,

$$
z = (x + yi)
$$

Then,

$$
|(x + yi) - 4| \ge |x + yi|
$$

Distances are positive, so,

$$
|(x - 4) + yi|^2 \ge |x + yi|^2
$$

Then,

$$
(x - 4)^2 + y^2 \ge x^2 + y^2
$$

$$
(x - 4)^2 \ge x^2
$$

$$
x^2 - 8x + 16 \ge x^2
$$

$$
16 \ge 8x
$$

$$
2 \ge x
$$

So, it's the numbers with real component less than or equal to 2.

This is not a domain because it's closed.
