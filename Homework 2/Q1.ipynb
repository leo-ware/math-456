Page 31: problem 5

**According to Sec. 10, the three cube roots of a nonzero complex number $z_0$ can be written $c_0$, $c_0ω_3$, $c_0ω_3^2$, where $c_0$ is the principal cube root of $z_0$ and,**

$$
\omega_3 = e^{2\pi i / 3}
$$

**Show that if $z_0 = −4\sqrt{2} + 4\sqrt{2}i$, then $c_0 = \sqrt{2}(1 + i)$ and the other two cube roots are,
in rectangular form, the numbers,**

$$
c_0ω_3 = {-(\sqrt{3} + 1) + (\sqrt{3} - 1)i \over \sqrt{2}}
$$

$$
c_0ω_3^2 = {(\sqrt{3} - 1) - (\sqrt{3} + 1)i \over \sqrt{2}}
$$


*Step 1: simplify $z_0$*

We have that,

$$
z_0 = -4\sqrt{2} + 4\sqrt{2}i
$$

So,

$$
|z_0| = \sqrt{(-4\sqrt{2})^2 + (4\sqrt{2})^2} = \sqrt{32 + 32} = \sqrt{64} = 8
$$

$$
\arg(z_0) = \arctan\left(\frac{4\sqrt{2}}{-4\sqrt{2}}\right) = \arctan(-1) = \frac{3\pi}{4}
$$

$$
z_0 = 8e^{3\pi i / 4}
$$

*Step 2: find the principal cube root*

We have that,

$$
c_0 = z_0^{1/3}
$$

So,

$$
c_0 = (8\exp(3\pi i / 4))^{1/3}
$$

$$
c_0 = 2\exp(\pi i / 4)
$$

$$
c_0 = \sqrt{2}(1 + i)
$$

So, $\sqrt{2}(1 + i)$ is the principal cube root of $z_0$.

*Step 3: show $c_0\omega_3$ is a cube root of $z_0$*

We have that,

$$
\omega_3 = e^{2\pi i / 3}
$$

$$
c_1 = c_0\omega_3 = {-(\sqrt{3} + 1) + (\sqrt{3} - 1)i \over \sqrt{2}}
$$

So,

$$
c_1 = 2\exp(\pi i / 4)e^{2\pi i / 3}
$$

$$
c_1 = 2\exp(\pi i(1/4 + 2/3))
$$

$$
c_1 = 2\exp(\pi i(11/ 12))
$$

So,

$$
c_1^3 = (2\exp(\pi i(11/ 12)))^3
$$

$$
c_1^3 = 8\exp(3\pi i(11/ 12))
$$

$$
c_1^3 = 8\exp(\pi i(11/ 4))
$$

$$
c_1^3 = 8\exp(\pi i(2 + 3/4))
$$

$$
c_1^3 = 8\exp(3\pi i / 4)
$$

$$
c_1^3 = z_0
$$


*Step 4: show $c_0\omega_3^2$ is a cube root of $z_0$*

We have that,

$$
c_2 = c_0\omega_3^2 = {(\sqrt{3} - 1) - (\sqrt{3} + 1)i \over \sqrt{2}}
$$

So,

$$
c_2 = 2\exp(\pi i / 4)\exp(2\pi i / 3)^2
$$

$$
c_2 = 2\exp(\pi i / 4)\exp(4\pi i / 3)
$$

$$
c_2 = 2\exp(\pi i (1/ 4 + 4/3))
$$

$$
c_2 = 2\exp(\pi i (19/ 12))
$$

$$
c_2^3 = (2\exp(\pi i (19/ 12)))^3
$$

$$
c_2^3 = 8\exp(3\pi i (19/ 12))
$$

$$
c_2^3 = 8\exp(\pi i (19/ 4))
$$

$$
c_2^3 = 8\exp(\pi i (4 + 3/4))
$$

$$
c_2^3 = 8\exp(3\pi i / 4)
$$

$$
c_2^3 = z_0
$$
