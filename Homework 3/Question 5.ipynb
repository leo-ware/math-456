**8. (a) Recall (Sec. 6) that if $z = x + iy$, then**

$$
x = {z + \bar{z} \over 2}
$$

$$
y = {z - \bar{z} \over 2i}
$$

**By *formally* applying the chain rule in calculus to a function $F(x, y)$ of two real variables, derive the expression**

$$
{\partial F \over \partial \bar{z}} =
{\partial F \over \partial x}{\partial x \over \partial \bar{z}} + {\partial F \over \partial y}{\partial y \over \partial \bar{z}} =
{1 \over 2}\left( {\partial F \over \partial x} + i{\partial F \over \partial y} \right)
$$

Let $z$ be a complex number such that,

$$
z = x + yi
$$

We have that,

$$
x = {z + \bar{z} \over 2}
$$

$$
y = {z - \bar{z} \over 2i}
$$

Recall the chain rule from calculus. Let $G$ be a real-valued function function of $n$ variables. Let $g_1(x)...g_n(x)$ be real-valued functions of $x$, and let $g(x)$ be the real vector-valued function,

$$
g(x) = \begin{bmatrix} g_1(x)\\\vdots\\g_n(x) \end{bmatrix}
$$

Then,

$$
{\partial \over \partial \phi}G(g_1(\phi)...g_n(\phi)) = (\nabla G)^T{\partial \over \partial \phi}g(\phi)
$$

Appying this identity, we get,

$$
{\partial \over \partial \bar{z}}F(x(\bar{z}), y(\bar{z})) =
(\nabla_{\bar{z}} F(x(\bar{z}), y(\bar{z})))^T
{\partial \over \partial \bar{z}} \left(\begin{bmatrix} x(\bar{z})\\y(\bar{z}) \end{bmatrix}\right)
$$

But,
$$
\nabla F =
\begin{bmatrix} {\partial F /\partial x}\\ {\partial F / y} \end{bmatrix}
$$

And,

$$
{\partial \over \partial \bar{z}} \left(\begin{bmatrix} x(\bar{z})\\y(\bar{z}) \end{bmatrix}\right) =
\begin{bmatrix} {\partial x /\partial \bar{z}}\\ {\partial y /\partial \bar{z}} \end{bmatrix}
$$

So,

$$
{\partial \over \partial \bar{z}}F(x(\bar{z}), y(\bar{z})) =
\begin{bmatrix} {\partial F /\partial x}\\ {\partial F / y} \end{bmatrix}^T
\begin{bmatrix} {\partial x /\partial \bar{z}}\\ {\partial y /\partial \bar{z}} \end{bmatrix}
$$

$$
{\partial \over \partial \bar{z}}F(x(\bar{z}), y(\bar{z})) =
{\partial F \over \partial x}{\partial x \over \partial\bar{z}} +
{\partial F \over \partial y}{\partial y \over \partial\bar{z}}
$$

Evaluating the partial derivatives, we get,

$$
{\partial x \over \partial \bar{z}} = {\partial \over \partial \bar{z}}{z + \bar{z} \over 2}
$$

$$
{\partial y \over \partial \bar{z}} = {\partial \over \partial \bar{z}}{z - \bar{z} \over 2i}
$$

So,

$$
{\partial x \over \partial \bar{z}} = {1 \over 2} \left( {\partial \over \partial \bar{z}}z + {\partial \over \partial \bar{z}}\bar{z} \right)
$$

$$
{\partial y \over \partial \bar{z}} = {1 \over 2} \left( {\partial \over \partial \bar{z}}z - {\partial \over \partial \bar{z}}\bar{z} \right)
$$

Evaluate,

$$
{\partial \bar{z} \over \partial \bar{z}} = 1
$$

Evaluate,

$$
{\partial z \over \partial \bar{z}} = 0
$$

This is the key step. We are thinking here of $z$ and $\bar{z}$ as independent variables, as an alternative coordinate system to the x/y system, rather than thinking of $z$ as a function of $\bar{z}$. So, the derivative is zero rather than being undefined.

Then,

$$
{\partial x \over \partial \bar{z}} = {1 \over 2} \left( 0 + 1 \right)
$$

$$
{\partial y \over \partial \bar{z}} = {1 \over 2i} \left( 0 - 1 \right)
$$

Then,

$$
{\partial x \over \partial \bar{z}} = {1 \over 2}
$$

$$
{\partial y \over \partial \bar{z}} = -{1 \over 2i}
$$

The multiplicative inverse of $i$ is $-i$, so,

$$
{\partial y \over \partial \bar{z}} = {i \over 2}
$$

Plug these back into the expression for derivative of $F$.

$$
{\partial \over \partial \bar{z}}F(x(\bar{z}), y(\bar{z})) =
{\partial F \over \partial x}{1 \over 2} +
{\partial F \over \partial y}{i \over 2}
$$

Then,

$$
{\partial \over \partial \bar{z}}F(x(\bar{z}), y(\bar{z})) =
{1 \over 2} \left(
{\partial F \over \partial x} +
i{\partial F \over \partial y}
\right)
$$

So,

$$
{\partial \over \partial \bar{z}}F(x(\bar{z}), y(\bar{z})) =
{\partial F \over \partial x}{\partial x \over \partial\bar{z}} +
{\partial F \over \partial y}{\partial y \over \partial\bar{z}} =
{1 \over 2} \left(
{\partial F \over \partial x} +
i{\partial F \over \partial y}
\right)
$$

**(b) Define the operator**

$$
{\partial \over \partial \bar{z}} = {1 \over 2}\left( {\partial \over \partial x} + i{\partial \over \partial y} \right)
$$

**suggested by part (a), to show that if the first-order partial derivatives of the real and imaginary components of a function $f (z)= u(x, y) + iv(x, y)$ satisfy the Cauchy–Riemann equations, then**

$$
{\partial f \over \partial \bar{z}} = {1 \over 2}[(u_x - v_y) + i(u_y + v_x)] = 0
$$

**Thus derive the complex form $∂f/∂\bar{z} = 0$ of the Cauchy–Riemann equations.**

Let $f(z)$ be a complex values function. Let $u(z)$ and $v(z)$ be real-valued functions such that, $f(z) = u(x, y) + iv(x, y)$. Consider the partial derivative of $f$ with respect to $\bar{z}$. By the above,

$$
{\partial f \over \partial \bar{z}} = {1\over 2}\left( {\partial f \over \partial x} + i{\partial f \over \partial y} \right)
$$

By the chain rule,

$$
{\partial f \over \partial x} = {\partial \over \partial x}u(x, y) + i{\partial \over \partial x}v(x, y)
$$

$$
{\partial f \over \partial y} = {\partial \over \partial y}u(x, y) + i{\partial \over \partial y}v(x, y)
$$

Then,

$$
{\partial f \over \partial \bar{z}} = {1\over 2}\left(
    \left( {\partial \over \partial x}u(x, y) + i{\partial \over \partial x}v(x, y) \right) +
    i\left( {\partial \over \partial y}u(x, y) + i{\partial \over \partial y}v(x, y) \right)
\right)
$$

$$
{\partial f \over \partial \bar{z}} = {1\over 2}\left(
    u_x + iv_x  + iu_y - v_y
\right)
$$

$$
{\partial f \over \partial \bar{z}} = {1\over 2}\left(
    (u_x - v_y) + i(u_y  + v_x)
\right)
$$

Suppose $f$ satisfies the Cauchy-Riemann equations. Then,

$$
u_x = v_y
$$

$$
u_y = -v_x
$$

Then,

$$
u_x - v_y = 0
$$

$$
u_y + v_x = 0
$$

$$
{1\over 2}\left(
    (u_x - v_y) + i(u_y  + v_x)
\right) = 0
$$

Then,

$$
{\partial f \over \partial \bar{z}} = {1\over 2}\left(
    (u_x - v_y) + i(u_y  + v_x)
\right) = 0
$$
