
**Consider the function $f : \mathbb{C} \to \mathbb{C}$ given by,**

$$
f(z) = \begin{cases}
{(y - 1)^3 - ix^3 \over x^2 + (y - 1)^2} & \text{if } z \ne i \\
0 & \text{if } z = i
\end{cases}
$$

**Prove that the Cauchy-Riemann equations hold at $z = i$. Is $f$ differentiable at $z = i$? If so, find the derivative.**

Let,

$$
u(x, y) = \begin{cases}
{(y - 1)^3 \over x^2 + (y - 1)^2} & \text{if } z \ne i \\
0 & \text{if } z = i
\end{cases}
$$

$$
v(x, y) = \begin{cases}
{-x^3 \over x^2 + (y - 1)^2} & \text{if } z \ne i \\
0 & \text{if } z = i
\end{cases}
$$

Then, $u$ and $v$ are real-valued functions such that,

$$
f = u + vi
$$

We are interested in the partial derivatives at $z = i$.

$$
u_x(0, 1) = \lim_{\Delta x \to 0} {u(0 + \Delta x, 1) - u(0, 1) \over x}
$$

$$
u_x(0, 1) = \lim_{x \to 0} {
    {((1) - 1)^3 \over (\Delta x)^2 + ((1) - 1)^2} - 0
\over \Delta x}
$$

$$
u_x(0, 1) = 0
$$

$$
u_y(0, 1) = \lim_{\Delta y \to 0} {u(0, \Delta y + 1) - u(0, 1) \over \Delta y}
$$

$$
u_y(0, 1) = \lim_{\Delta y \to 0} {
    {((\Delta y + 1) - 1)^3 \over (0)^2 + ((\Delta y + 1) - 1)^2}
    - 0
\over \Delta y}
$$

$$
u_y(0, 1) = \lim_{\Delta y \to 0} {
    \left( {\Delta y^3 \over \Delta y^2} \right)
\over \Delta y}
$$

$$
u_y(0, 1) = \lim_{\Delta y \to 0} {\Delta y^3 \over \Delta y^3}
$$

$$
u_y(0, 1) = 1
$$

$$
v_x(0, 1) = \lim_{\Delta x \to 0} {
    v(\Delta x + 0, 1) + v(0, 1)
\over \Delta x}
$$

$$
v_x(0, 1) = \lim_{\Delta x \to 0} {
    {-(\Delta x + 0)^3 \over (\Delta x + 0)^2 + ((1) - 1)^2} + 0
\over \Delta x}
$$

$$
v_x(0, 1) = \lim_{\Delta x \to 0} {
    {-\Delta x^3 \over \Delta x^2}
\over \Delta x}
$$

$$
v_x(0, 1) = -1
$$

$$
v_x(0, 1) = \lim_{\Delta y \to 0} {
    {-(0)^3 \over (0)^2 + ((\Delta y + 1) - 1)^2} + 0
\over \Delta y}
$$

$$
v_x(0, 1) = 0
$$

So,

$$
u_x = v_y = 0
$$

$$
u_y = -v_x = 1
$$

The Cauchy-Riemann equations are satisfied at $z = i$.

The derivative of $f$ at $z = i$ is given by,

$$
f'(0, 1) = \lim_{\Delta x, \Delta y \to 0} {f(\Delta x + 0, \Delta y + 1) - f(0, 1) \over \Delta x + i\Delta y}
$$

$$
f'(0, 1) = \lim_{\Delta x, \Delta y \to 0} {
{((\Delta y + 1) - 1)^3 - i\Delta x^3 \over \Delta x^2 + ((\Delta y + 1) - 1)^2}
- 0
\over \Delta x + i\Delta y}
$$

$$
f'(0, 1) = \lim_{\Delta x, \Delta y \to 0} {
\left( {\Delta y^3 - i\Delta x^3 \over \Delta x^2 + \Delta y^2} \right)
\over \Delta x + i\Delta y}
$$

$$
f'(0, 1) = \lim_{\Delta x, \Delta y \to 0}
{\Delta y^3 - i\Delta x^3 \over (\Delta x^2 + \Delta y^2)(\Delta x + i\Delta y)}
$$

Consider the approach $\Delta x \to 0$ while $\Delta y = 0$

$$
f'(0, 1) = \lim_{\Delta x\to 0}
{(0)^3 - i\Delta x^3 \over (\Delta x^2 + (0)^2)(\Delta x + i(0))}
$$

$$
f'(0, 1) = \lim_{\Delta x\to 0}
{- i\Delta x^3 \over \Delta x^3}
$$

$$
f'(0, 1) = \lim_{\Delta x\to 0}
- i
$$

$$
f'(0, 1) = -i
$$

Consider the approach $\Delta y \to 0$ while $\Delta x = 0$

$$
f'(0, 1) = \lim_{\Delta y \to 0}
{\Delta y^3 - i(0)^3 \over ((0)^2 + \Delta y^2)((0) + i\Delta y)}
$$

$$
f'(0, 1) = \lim_{\Delta y \to 0}
{\Delta y^3 \over i\Delta y^3}
$$

$$
f'(0, 1) = \lim_{\Delta y \to 0}
{1 \over i}
$$

$$
f'(0, 1) = {1 \over i}
$$

But, limits are unique. So, $f'(0, 1)$ does not exist.
