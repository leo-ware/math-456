**Where is the function $f(z) = z\operatorname{Im}(z^2) - i(\bar{z} + 1)^2$ differentiable?**

$$
f(z) = z\operatorname{Im}(z^2) - i(\bar{z} + 1)^2
$$

So,

$$
f(z) = (x + yi)2xy - i\bar{z}^2 -2i\bar{z} -i
$$

$$
f(z) = (x + yi)2xy - i((x^2 - y^2) - 2xyi) - 2i(x - yi) -i
$$

$$
f(z) = 2x^2y + 2xy^2i - ix^2 + y^2i - 2xy - 2ix -2y - i
$$

$$
f(z) = (2x^2y - 2xy -2y) + i(2xy^2 - x^2 + y^2 - 2x - 1)
$$


So,

$$
u = 2x^2y - 2xy - 2y
$$

$$
v = 2xy^2 - x^2 + y^2 - 2x - 1
$$

So,

$$
u_x = 4xy - 2y
$$

$$
u_y = 2x^2 - 2x - 2
$$

And,

$$
v_x = 2y^2 - 2x - 2
$$

$$
v_y = 4xy + 2y
$$

The partials are continuous, so where the Cauchy-Riemann equations hold, the derivative will be defined. Checking the first condition, we get,

$$
u_x = v_y
$$

$$
4xy - 2y = 4xy + 2y
$$

$$
y = -y
$$

$$
y = 0
$$

Checking the second condition, we get,

$$
u_y = -v_x
$$

$$
2x^2 - 2x - 2 = -(2y^2 - 2x - 2)
$$

$$
2x^2 - 2x - 2 = -2y^2 + 2x + 2
$$

We already know the derivative is only defined when $y = 0$, so we can constrain the search to this region.

$$
2x^2 - 2x - 2 = 2x + 2
$$

$$
x^2 - 2x - 2 = 0
$$

Applying the quadratic formula, we get,

$$
x = {-(-2) \pm \sqrt{(-2)^2 - 4(1)(-2)} \over 2}
$$

$$
x = {2 \pm \sqrt{4 + 8} \over 2}
$$

$$
x = {2 \pm 2\sqrt{3} \over 2}
$$

$$
x = 1 \pm \sqrt{3}
$$

So, the derivative is defined at the real numbers $1 + \sqrt{3}$ and $1 - \sqrt{3}$.
