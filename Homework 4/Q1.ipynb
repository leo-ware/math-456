**(a) $f(z) = {1 \over z^4}$ where $z \ne 0$**

$$
f(z) = {1 \over z^4}
$$

$$
f(z) = z^{-4}
$$

Let,

$$
z = \rho e^{i\theta}
$$

Then,

$$
f(z) = (\rho e^{i\theta})^{-4}
$$

$$
f(z) = \rho^{-4} e^{-4i\theta}
$$

$$
f(z) = \rho^{-4} (\cos(-4\theta) + i\sin(-4\theta))
$$

Let, $u$ and $v$ be real-valued functions such that,

$$
f(z) = u(z) + iv(z)
$$

Then,

$$
u(z) = \rho^{-4} \cos(-4\theta)
$$

$$
v(z) = \rho^{-4} \sin(-4\theta)
$$

Take the partials.

$$
u_\rho(z) = {d \over d\rho} \rho^{-4} \cos(-4\theta) = -4\rho^{-5} \cos(-4\theta) 
$$

$$
u_\theta(z) = {d \over d\theta} \rho^{-4} \cos(-4\theta) = 4\rho^{-4} \sin(-4\theta)
$$

$$
v_\rho(z) = {d \over d\rho} \rho^{-4} \sin(-4\theta) = -4\rho^{-5} \sin(-4\theta)
$$

$$
v_\theta(z) = {d \over d\theta} \rho^{-4} \sin(-4\theta) = -4\rho^{-4} \cos(-4\theta)
$$

When $z \ne 0$, $\rho \ne 0$. In this case, each of the partials is the ratio of a continuous trig function and a nonzero polynomial, which is continuous. The set of continuous functions is closed under nonzero division, so the partials are continuous.

Then,

$$
\rho u_\rho = (\rho)(-4)\rho^{-5} \cos(-4\theta)
$$

$$
\rho u_\rho = -4\rho^{-4} \cos(-4\theta)
$$

$$
\rho u_\rho = v_\theta
$$

Also,

$$
-\rho v_r = (-\rho)(-4)\rho^{-5} \sin(-4\theta)
$$

$$
-\rho v_r = 4\rho^{-4} \sin(-4\theta)
$$

$$
-\rho v_r = u_\theta
$$

So, the polar Cauchy-Riemann is satisfied. We established earlier that the partials are continuous under the assumption that $z \ne 0$. So, when $z \ne 0$, the derivative is defined.

Specifically, the derivative is given by,

$$
f'(z) = e^{-i\theta}(u_r + iv_r)
$$

$$
f'(z) = e^{-i\theta}((-4\rho^{-5} \cos(-4\theta)) + i(-4\rho^{-5} \sin(-4\theta)))
$$

$$
f'(z) = e^{-i\theta}(-4)\rho^{-5}(\cos(-4\theta) + i\sin(-4\theta))
$$

By Euler's identity,

$$
f'(z) = -4\rho^{-5}e^{-i\theta}e^{-4i\theta}
$$

So,

$$
f'(z) = -4\rho^{-5}e^{-5i\theta}
$$

**(b) $f(z) = e^{-\theta}\cos(\ln r) + ie^{-\theta}\sin(\ln r)$ where $r > 0, 0 < \theta < 2\pi$**

Let $u$ and $v$ be real-valued functions such that $f(z) = u(z) + v(z)i$. Then,

$$
u = e^{-\theta}\cos(\ln r)
$$

$$
v = e^{-\theta}\sin(\ln r)
$$

For these to be defined, we require that $r \ne 0$. We further stipulate that $0 \lt \theta\lt 2\pi$.

Take the partials.

$$
u_r = {d \over d\rho} e^{-\theta}\cos(\ln r) = -e^{-\theta}\sin(\ln r)r^{-1}
$$

$$
u_\theta = {d \over d\theta} e^{-\theta}\cos(\ln r) = -e^{-\theta}\cos(\ln r)
$$

$$
v_r = e^{-\theta}\sin(\ln r) = e^{-\theta}\cos(\ln r)r^{-1}
$$

$$
v_\theta = {d \over d\theta} e^{-\theta}\sin(\ln r) = -e^{-\theta}\sin(\ln r)
$$

Each of the partials is a product of a continuous trig function and an exponential, which is continuous. The set of continuous functions is closed under multiplication. So, the partials are continuous.

Consider the quantity $ru_r$.

$$
ru_r = -re^{-\theta}\sin(\ln r)r^{-1}
$$

$$
ru_r = -e^{-\theta}\sin(\ln r)
$$

$$
ru_r = v_\theta
$$

Consider the quantity $-rv_r$.

$$
-rv_r = (-r)e^{-\theta}\cos(\ln r)r^{-1}
$$

$$
-rv_r = -e^{-\theta}\cos(\ln r)
$$

$$
-rv_r = u\theta
$$

So, the Cauchy-Riemann equations are satisfied. We already established that the partials are continuous for all $z$ subject to $r \ne 0$ and $0 \lt \theta \lt 2\pi$. This suffices to show that the derivative is defined on this range. The derivative is given by,

$$
f'(z) = e^{-i\theta}(u_r + iv_r)
$$

$$
f'(z) = e^{-i\theta}((-e^{-\theta}\sin(\ln r)r^{-1}) + i(e^{-\theta}\cos(\ln r)r^{-1}))
$$

Multiply by $i$ and $-i$.

$$
f'(z) = i(-i)e^{-i\theta}r^{-1}(-e^{-\theta}\sin(\ln r) + ie^{-\theta}\cos(\ln r))
$$

$$
f'(z) = ie^{-i\theta}r^{-1}(-(-i)e^{-\theta}\sin(\ln r) + i(-i)e^{-\theta}\cos(\ln r))
$$

$$
f'(z) = i(e^{i\theta}r)^{-1}(e^{-\theta}\cos(\ln r) + ie^{-\theta}\sin(\ln r))
$$

$$
f'(z) = i{e^{-\theta}\cos(\ln r) + ie^{-\theta}\sin(\ln r) \over e^{i\theta}r}
$$

$$
f'(z) = i{f(z) \over z}
$$
