```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fn(z):
    return z**3

def fn_diff_norm(z):
    diff = fn(z) - z
    mod = np.abs(diff)
    mod[mod == 0] = 1
    return diff / mod
```

```python
x = y = np.linspace(-10, 10, 20)
xm, ym = np.meshgrid(x, y)
points = (xm + ym*1j).flatten()
```

```python
diffs = fn_diff_norm(points)

plt.quiver(
    np.real(points),
    np.imag(points),
    np.real(diffs),
    np.imag(diffs)
)

xlim = plt.xlim()
ylim = plt.ylim()
plt.vlines(0, ylim[0], ylim[1], color="black")
plt.hlines(0, xlim[0], xlim[1], color="black")
plt.xlim(xlim)
plt.ylim(ylim)

plt.show()
```
